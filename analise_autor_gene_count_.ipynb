```python
#Análise descritiva:
import pandas as pd

# Carregar o dataset (substitua 'nome_do_arquivo.csv' pelo nome do seu arquivo)
df = pd.read_csv('autor_gene_count2.csv', names=['Author', 'Article_Count'])

# Remover linhas que contenham valores não numéricos na coluna 'Artigos'
df = df[pd.to_numeric(df['Article_Count'], errors='coerce').notnull()]

# Converter a coluna 'Artigos' para valores inteiros
df['Article_Count'] = df['Article_Count'].astype(int)

# Estatísticas descritivas
media_Articles = df['Article_Count'].mean()
mediana_Articles = df['Article_Count'].median()
desvio_padrao_Articles = df['Article_Count'].std()

print(f"Average number of articles per author: {media_Articles}")
print(f"Median number of articles per author: {mediana_Articles}")
print(f"Standard deviation of articles per author: {desvio_padrao_Articles}")
```

```python
#Top autores por número de artigos:

# Ordenar o DataFrame por número de artigos em ordem decrescente
df_sorted = df.sort_values(by='Article_Count', ascending=False)

# Mostrar os top N autores
top_n = 10  # Defina o número de autores que deseja mostrar
top_autores = df_sorted.head(top_n)

print(f"Top {top_n} Author by number of articles:")
print(top_autores)
```

```python
#Histograma:
import matplotlib.pyplot as plt

# Aumentar o tamanho da figura
plt.figure(figsize=(10, 6))  # Defina o tamanho desejado (largura x altura)

# Criar o histograma
hist = plt.hist(df['Article_Count'], bins=20, edgecolor='k')
plt.xlabel('Numbers of articles')
plt.ylabel('Frequencies')
plt.title('Distribution of the Number of Articles per Author')

# Adicionar os valores acima de cada barra
for i in range(len(hist[0])):
    count = hist[0][i]
    x = hist[1][i]
    y = hist[0][i]
    plt.text(x + (hist[1][1] - hist[1][0]) / 2, y, str(count), ha='center', va='bottom')

# Mostrar o gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np  # Importar numpy para cálculos estatísticos

# Aumentar o tamanho da figura
plt.figure(figsize=(10, 6))

# Criar o histograma
hist, bins, _ = plt.hist(df['Article_Count'], bins=20, edgecolor='k')

# Calcular estatísticas
max_value = np.max(df['Article_Count'])
min_value = np.min(df['Article_Count'])
mean_value = np.mean(df['Article_Count'])

# Adicionar os valores acima de cada barra
for i in range(len(hist)):
    count = hist[i]
    x = bins[i]
    y = hist[i]
    plt.text(x + (bins[1] - bins[0]) / 2, y, str(int(count)), ha='center', va='bottom')

# Mostrar o gráfico
plt.xlabel('Numbers of articles')
plt.ylabel('Frequencies')
plt.title('Distribution of the Number of Articles per Author')
plt.axvline(mean_value, color='r', linestyle='dashed', linewidth=1, label=f'Mean: {mean_value:.2f}')
plt.legend()
plt.show()

# Exibir estatísticas
print(f"Valor Máximo: {max_value}")
print(f"Valor Mínimo: {min_value}")
print(f"Média: {mean_value}")
```
